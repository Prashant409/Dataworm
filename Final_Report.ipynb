## Trend of Orders/Different dates Pre-Orders with Date:

```python
#@title Trend of Orders/Pre-Orders with Date:
plt.figure(figsize=(12, 6))

output = df.groupby('date_month_invoice')['id_order'].agg(['count']).round()
def get_difference(row):
    date_month_delivery = pd.to_datetime(row['date_month_delivery'])
    date_month_invoice = pd.to_datetime(row['date_month_invoice'])
    delta = date_month_delivery - date_month_invoice
    return delta.days


rslt_df = df.loc[(df['preorder'] == 1) & (df.apply(get_difference, axis=1) > 0)] 


rslt_df = rslt_df.groupby('date_month_invoice')['id_order'].agg(['count'])

output = output.sort_values('date_month_invoice')
rslt_df = rslt_df.sort_values('date_month_invoice')

plt.plot(output.index, output['count'], label='Orders', marker='o', color='#4c78a8')

plt.plot(rslt_df.index, rslt_df['count'], label='Pre-Orders', marker='o', linestyle='--', color='red')

plt.xlabel('Invoice Date',fontsize=12,fontweight='bold',color='grey')
plt.ylabel('Total Orders',fontsize=12,fontweight='bold',color='grey')
plt.legend(fontsize=12 ,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Total Orders/Pre-Orders by Invoice Date', fontsize=14,fontweight='bold',color='grey')
plt.grid(linestyle='--', alpha=0.7)

for x, y in zip(output.index, output['count']):
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=12,color='red')

for x, y in zip(rslt_df.index, rslt_df['count']):
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=12,color='blue')  
    
   
sns.despine()
plt.show()
```
